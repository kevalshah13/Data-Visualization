```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv(r"\Data-Visualization\exams.csv")
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
value=data.groupby(['gender']).mean()
colors = ['orange', 'cyan', 'green']
ax=[ax1,ax2,ax3]

for i in range(len(value.columns)):
    ax[i].bar(value.index, value[value.columns[i]],color=colors[i])
    ax[i].set_xlabel("Gender")
    ax[i].set_ylabel(value.columns[i])
plt.suptitle("Score Distribution")

plt.show()
```
